### Code to collect the images of the person: 

```python
import os
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import numpy as np
import random
import shutil
```

```python
path = r"C:\Users\Muntasir\Desktop\Assignment3\coco_minitrain_25k\coco_minitrain_25k"
```

```python
label_text_files  = os.listdir(path+"\\labels\\train2017")
```

```python
person_image_file_names = []

for i in label_text_files:
    if list(pd.read_csv(path+"\\labels\\train2017\\"+i,delimiter = " ", header = None)[0]).count(0)>0:
        person_image_file_names.append(i)
```

```python
image_path = "C:\\Users\\Muntasir\\Desktop\\Assignment3\\coco_minitrain_25k\\coco_minitrain_25k\\images\\train2017\\"
```

```python
len(person_image_file_names)
```

```python
for k in random.sample(range(0, 13629), 10):
    image = mpimg.imread(image_path+person_image_file_names[k].split(".")[0]+".jpg")
    plt.imshow(image)
    plt.show()
```

```python
src = r"C:\Users\Muntasir\Desktop\Assignment3\coco_minitrain_25k\coco_minitrain_25k\images\train2017"
dst = r"C:\Users\Muntasir\Desktop\Assignment3\coco_minitrain_25k\coco_minitrain_25k\images\class_person"
```
